# 1. 영화랭킹(평점순)에서 장르별 30개씩 스크래핑하기

```python
import datetime
today = datetime.date.today()
yesterday = today - datetime.timedelta(1)  
date = yesterday.strftime('%Y%m%d')

print(date)
```

```python
# 랭킹 1~50위까지 영화제목, 상세페이지 크롤링
import requests
from bs4 import BeautifulSoup
import bs4
from urllib.parse import urljoin
import re

main_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date='+ date

movie_list_2 = []
for i in [1,2,4,5,6,7,11,13,15,16,18,19]:
    movie_genre_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20200818&tg='+str(i)
    html = requests.get(movie_genre_url).text
    soup = BeautifulSoup(html, 'html.parser')

    a_tag_list = soup.select('a[href*=basic.nhn]')


    for idx,a_tag in enumerate(a_tag_list,1):
        if idx > 30:
            break
        movie_dict = {}
        title = a_tag.text.strip()
        link = urljoin(main_url, a_tag['href'])
        
        movie_dict['genre_code'] = i
        movie_dict['id'] = idx
        movie_dict['title'] = title
        movie_dict['url'] = link
        # scorre_url 은 평점을 스크래핑하기 위한 페이지
        movie_dict['score_url'] = link.replace('basic', 'pointWriteFormList')
        movie_list_2.append(movie_dict)
        #print(idx, title, link)
        
movie_list_2
```

```python
movie_code_list = []
for i in [1,2,4,5,6,7,11,13,15,16,18,19]:
    movie_genre_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=pnt&date=20200818&tg='+str(i)
    html = requests.get(movie_genre_url).text
    soup = BeautifulSoup(html, 'html.parser')
    
    a_tag_list = soup.select('a[href*=basic.nhn]')
    
    for idx,a_tag in enumerate(a_tag_list,1):
        if idx > 30:
            break
        #print(a_tag['href'])
        movie_code = re.findall("\d+", a_tag['href'])
        #movie_code = int(filter(str.isdigit, a_tag['href']))
        for i in movie_code:
            movie_code_list.append(i)

print(len(movie_code_list))    
movie_code_list
```

```python
movie_info_lists_2 = []

for movie in movie_list_2:
    html_1 = requests.get(movie['url']).text
#     html_2 = requests.get(movie['score_url']).text
    soup_1 = BeautifulSoup(html_1,'html.parser')
#     soup_2 = BeautifulSoup(html_2,'html.parser')
    
    movie_info_dict = {}
    
    # 타이틀
    title = soup_1.select_one('h3.h_movie a[href*=basic.nhn]')
    if title == None:
        movie_info_dict['title'] = None
    else:
        movie_info_dict['title'] = title.text.strip().replace(',', '')
    
    # 네티즌 평점
    score = soup_1.select_one('div.netizen_score .star_score em')
    if score == None:
        movie_info_dict['score'] = None
    else:
        movie_info_dict['score'] = score.text.strip()
#     if score == '관람객 평점 없음':
#         movie_info_dict['score'] = None
#     elif score == '평점 - 총 10점 중':
#         movie_info_dict['score'] = None
#     else:
#         movie_info_dict['score'] = re.findall("(\d+\.+\d+)", score)[0]
        

    #movie_info_dict['genre'] = re.sub(r"\s+", "", soup_1.select_one('dl.info_spec span').text.strip())
    
    
    # 개봉국가
    nation = soup_1.select_one('dl.info_spec span a[href*=nation]')
    if nation == None:
        movie_info_dict['nation'] = None
    else:
        movie_info_dict['nation'] = nation.text.strip()
    
    
    # 감독
    director = soup_1.select_one('dl.info_spec a[href*=code]')
    if director == None:
        movie_info_dict['director'] = None
    else:
        movie_info_dict['director'] = director.text.strip()
    
    
    # movie_info_dict['actor'] = soup_1.find('dl',attrs={"class":"info_spec"}).find_all("dd")[2].text.replace('더보기', '')
    
    # 누적관객수 
    audience = soup_1.select_one('p.count')
    if audience == None:
        movie_info_dict['audience'] = None
    else:
        movie_info_dict['audience'] = re.sub(r"(명)\([^()]*\)", "", audience.text.strip()).replace(',', '')
    
    # 관람객 평점 건수
    comment_count = soup_1.select_one('div.netizen_score .user_count em')
    if comment_count == None:
        movie_info_dict['comment_count'] = None
    else:
        movie_info_dict['comment_count'] = comment_count.text.strip().replace(',', '')
    
    
    # 개봉일자
    release_year = soup_1.select_one('dl.info_spec a[href*=open]')
    if release_year == None:
        movie_info_dict['release_year'] = None
    else:
        movie_info_dict['release_year'] = release_year.text.strip()
    
    
    # score_sum = 관람객 평점 * 관람객 건수
#     score_num = float(re.findall("(\d+\.+\d+)", score)[0])
#     comment_count_num = int(soup_2.select_one('.total em').text.strip().replace(',', ''))

    movie_info_dict['score_sum'] = 0

    # 출거리
    plot = soup_1.select_one('.con_tx')
    if plot == None:
        movie_info_dict['plot'] = None
    else: 
        movie_plot = " ".join(re.sub('|  |\t|\r|\n|', '', plot.text).split())
        movie_info_dict['plot'] = movie_plot.replace(',',"")

    movie_info_lists_2.append(movie_info_dict)

print(movie_info_lists_2)
```

```python
movie_info_lists = []

for movie in movie_list:
    html_1 = requests.get(movie['url']).text
    html_2 = requests.get(movie['score_url']).text
    soup_1 = BeautifulSoup(html_1,'html.parser')
    soup_2 = BeautifulSoup(html_2,'html.parser')
    
    movie_info_dict = {}
    
    movie_info_dict['title'] = soup_1.select_one('h3.h_movie a[href*=basic.nhn]').text.strip()
    
    score = soup_1.select_one('div.star_score span.st_on').text.strip()
    if score == '관람객 평점 없음':
        movie_info_dict['score'] = None
    elif score == '평점 - 총 10점 중':
        movie_info_dict['score'] = None
    else:
        movie_info_dict['score'] = re.findall("(\d+\.+\d+)", score)[0]
        

    #movie_info_dict['genre'] = re.sub(r"\s+", "", soup_1.select_one('dl.info_spec span').text.strip())
    
    movie_info_dict['nation'] = soup_1.select_one('dl.info_spec span a[href*=nation]').text.strip()
    
    movie_info_dict['director'] = soup_1.select_one('dl.info_spec a[href*=code]').text.strip()
    
    # movie_info_dict['actor'] = soup_1.find('dl',attrs={"class":"info_spec"}).find_all("dd")[2].text.replace('더보기', '')
    
    # 누적관객수 
    if soup_1.select_one('p.count') == None:
        movie_info_dict['audience'] = None
    else:
        movie_info_dict['audience'] = re.sub(r"(명)\([^()]*\)", "", soup_1.select_one('p.count').text.strip()).replace(',', '')
    
    # 관람객 평점 건수
    movie_info_dict['comment_count'] = soup_2.select_one('.total em').text.strip().replace(',', '')
    
    
    # 개봉일자
    movie_info_dict['release_year'] = soup_1.select_one('dl.info_spec a[href*=open]').text.strip()
    
    
    # score_sum = 관람객 평점 * 관람객 건수
#     score_num = float(re.findall("(\d+\.+\d+)", score)[0])
#     comment_count_num = int(soup_2.select_one('.total em').text.strip().replace(',', ''))

    movie_info_dict['score_sum'] = 0
 
    plot = soup_1.select_one('.con_tx')
    if plot == None:
        movie_info_dict['plot'] = None
    else: 
        movie_plot = " ".join(re.sub('|  |\t|\r|\n|', '', plot.text).split())
        movie_info_dict['plot'] = movie_plot.replace(',',"")

    movie_info_lists.append(movie_info_dict)

print(movie_info_lists)
```

```python
# 판다스로 데이터프레임 만들고 그 안에 series 데이터 넣기


import pandas as pd

data_df = pd.DataFrame(columns=['title', 'score', 'nation', 'director', 'audience', 'plot', 'score_sum', 'comment_count', 'release_year'])

for data in movie_info_lists:
    series_obj = pd.Series(data)
    data_df = data_df.append(series_obj, ignore_index=True)
    
data_df
```

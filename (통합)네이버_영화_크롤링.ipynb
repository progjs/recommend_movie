# 1. 영화 랭킹 1~50위까지 영화제목, URL 크롤링

```python
# 랭킹 1~50위까지 영화제목, 상세페이지 크롤링
import requests
from bs4 import BeautifulSoup
import bs4
from urllib.parse import urljoin
import re

main_url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
html = requests.get(main_url).text
soup = BeautifulSoup(html, 'html.parser')

a_tag_list = soup.select('a[href*=basic.nhn]')

movie_list = []
for idx,a_tag in enumerate(a_tag_list,1):
    if idx > 50:
        break
    movie_dict = {}
    title = a_tag.text.strip()
    link = urljoin(main_url, a_tag['href'])
    
    movie_dict['title'] = title
    movie_dict['url'] = link
    movie_list.append(movie_dict)
    #print(idx, title, link)
movie_list
```

## 4. 크롤링데이터 데이터베이스에 넣기 & 판다스

```python
movie_info_lists = []

for movie in movie_list:
    html = requests.get(movie['url']).text
    soup = BeautifulSoup(html,'html.parser')
    
    movie_info_dict = {}
    
    movie_info_dict['title'] = soup.select_one('h3.h_movie a[href*=basic.nhn]').text.strip()
    
    score = soup.select_one('div.star_score span.st_on').text.strip()
    if score == '관람객 평점 없음':
        movie_info_dict['score'] = None
    elif score == '평점 - 총 10점 중':
        movie_info_dict['score'] = None
    else:
        movie_info_dict['score'] = re.findall("(\d+\.+\d+)", score)[0]
        

    movie_info_dict['genre'] = re.sub(r"\s+", "", soup.select_one('dl.info_spec span').text.strip())
    
    movie_info_dict['nation'] = soup.select_one('dl.info_spec span a[href*=nation]').text.strip()
    
    movie_info_dict['director'] = soup.select_one('dl.info_spec a[href*=code]').text.strip()
    
    movie_info_dict['actor'] = soup.find('dl',attrs={"class":"info_spec"}).find_all("dd")[2].text.replace('더보기', '')
    
    if soup.select_one('p.count') == None:
        movie_info_dict['audience'] = None
    else:
        movie_info_dict['audience'] = re.sub(r"(명)\([^()]*\)", "", soup.select_one('p.count').text.strip()).replace(',', '')


    
    plot = soup.select_one('.con_tx')
    if plot == None:
        movie_info_dict['plot'] = None
    else: 
        movie_plot = " ".join(re.sub('|  |\t|\r|\n|', '', plot.text).split())
        movie_info_dict['plot'] = movie_plot

    movie_info_lists.append(movie_info_dict)

print(movie_info_lists)
```

```python
# 판다스로 데이터프레임 만들고 그 안에 series 데이터 넣기

import pandas as pd

data_df = pd.DataFrame(columns=['title', 'score', 'genre', 'nation', 'director', 'actor', 'audience', 'plot'])

for data in movie_info_lists:
    series_obj = pd.Series(data)
    data_df = data_df.append(series_obj, ignore_index=True)

data_df  
```

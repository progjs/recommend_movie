```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re

url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')

a_tag_list = soup.select('.title a[href*=basic.nhn]')

for idx, a_tag in enumerate(a_tag_list, 1):
    title = a_tag.text
    link = urljoin(url, a_tag['href'])
    print(idx, title, link)
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import re

movie_num = 189069
url = 'https://movie.naver.com/movie/bi/mi/basic.nhn?code='+str(movie_num)
html = requests.get(url).text
soup = BeautifulSoup(html)

title = soup.select_one('.wide_info_area a[href^=./basic.nhn]')
genre_tag_list = soup.select('.wide_info_area a[href*=genre]')
nation_tag_list = soup.select('.wide_info_area a[href*=nation]')
director_tag_list = soup.select('.info_spec dd:nth-of-type(2) p a')
actor_tag_list = soup.select('.info_spec dd:nth-of-type(3) p a')
count_tag_list = soup.select('.info_spec dd:nth-of-type(5) .count')
rate_tag_list = soup.select('#actualPointPersentBasic .st_on')

print(title.text)

print(genre_tag_list)

for nation in nation_tag_list:
    print(nation.text)
    
for director in director_tag_list:
    print(director.text)
    
for actor in actor_tag_list:
    print(actor.text)

for count in count_tag_list:
    print(re.sub('\((.*)\)', '', count.text))
    
for rate in rate_tag_list:
    print(re.sub('([ㄱ-힑 ])+', "", rate.text) + '점')
    
    
for i in range(1, 11):
    url2 = 'https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code='+str(movie_num)+'&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page='
    print('================='+str(i)+'페이지=================')
    urls2 = url2 + str(i)
    html = requests.get(urls2).text
    soup = BeautifulSoup(html, 'html.parser')
    comment_tag_list = soup.select('span[id^=_filtered_ment_]')
    
    
    for idx, comment_tag in enumerate(comment_tag_list, 1):
        comment = comment_tag.text.strip()
        no_space = re.sub('|  |\t|\n', '', comment)
        print(idx, no_space)
    
    if len(comment_tag_list) != 10:
        break;
    
```
